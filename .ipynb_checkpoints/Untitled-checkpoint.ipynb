```python
import matplotlib.pyplot as plt

# Define the layers of the neural network
layers = ['Input Layer\n(Color, Piece, Square)', 'Hidden Layer', 'Output Layer\nCritical Power Level']

# Define the connections between layers
connections = [('Input Layer\n(Color, Piece, Square)', 'Hidden Layer'), ('Hidden Layer', 'Output Layer\nCritical Power Level')]

# Create the figure and axis
fig, ax = plt.subplots(figsize=(6, 3))

# Draw the nodes (layers)
for i, layer in enumerate(layers):
    if i == 0:
        ax.text(i-0.1, 0.5, layer, ha='center', va='center', bbox=dict(boxstyle='round,pad=0.3', fc='lightgray', ec='gray'))
    else:
        ax.text(i, 0.5, layer, ha='center', va='center', bbox=dict(boxstyle='round,pad=0.3', fc='lightgray', ec='gray'))

# Draw the connections
for connection in connections:
    start_layer = layers.index(connection[0])
    end_layer = layers.index(connection[1])
    arrowprops = dict(arrowstyle='->', linewidth=1, edgecolor='gray')
    ax.annotate('', xytext=(start_layer, 0.5), xy=(end_layer, 0.5), arrowprops=arrowprops)

# Add the tanh activation function box
activation_box = plt.Rectangle((0.27, 0.15), 0.46, 0.7, facecolor='lightgray', edgecolor='gray')
ax.add_patch(activation_box)

# Set the axis limits and remove the ticks
ax.set_xlim(-0.7, len(layers) - 0.5)
ax.set_ylim(0, 1)
ax.axis('off')

# Add the tanh activation function label
ax.text(0.5, 0.5, 'tanh', ha='center', va='center')

# Set the title and show the diagram
ax.set_title('3-Layer Neural Network for Chess', fontweight='bold', fontsize=12)
plt.tight_layout()
plt.show()
```
